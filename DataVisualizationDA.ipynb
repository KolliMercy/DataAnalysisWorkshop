# Scatter plot:


```python
x=np.array([10,20,30,40])
y=np.array([50,55,32,67])
plt.scatter(x,y,color='m',marker='^')
plt.title('scatter plot')
plt.xlabel('xvalues')
plt.ylabel('yvalues')
plt.xlim(10,50)                    # to change the x,y limits
plt.ylim(10,100)
plt.show()
```

```python
x=np.array([10,20,30,40])
y=np.array([50,55,32,67])
plt.scatter(x,y,color='red')
plt.title('scatter')
plt.xlabel('x values')
plt.ylabel('y values')

for a,b in zip(x,y):
    plt.text(a,b,(str(a),str(b)))
    
plt.show()
```

### Subplots:


```python
x=np.array([10,20,30,40])
y=np.array([50,55,32,67])
plt.subplot(2,2,1)
plt.plot(x,y,'r')
plt.subplot(2,2,4)
plt.plot(x**8,y,'m')
plt.subplot(2,2,2)
plt.plot(x,y,'b')
plt.subplot(2,2,3)
plt.plot(x,y,'g')
```

### pie plot


```python
lang=['Kova','conda','see','python','NET']
count=[90,100,67,98,40]

plt.pie(count,labels=lang,explode=[0,0,0,0,0],autopct='%1.2f%%')
plt.show()
```

```python
from sklearn.datasets import load_iris
d=load_iris() 
df=pd.DataFrame(d.data,columns=d.feature_names)
df['target']=d.target
df.head(2)
```

```python
l=[]
for i in df['target']:
    if(i==0):
        l.append('setosa')
    elif(i==1):
        l.append('versicolor')
    elif(i==2):
        l.append('virginica')
df['classes']=l
df.sample(10)
```

```python
# catplot
sns.catplot(data=df)
```

```python
sns.catplot(data=df)
plt.xticks(rotation=25)           #specifices the rotation of xlimits names
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df)
plt.show()
```

```python
sns.catplot(x='classes',y='petal length (cm)',data=df)
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='swarm')
plt.show()
```

```python
sns.catplot(x='classes',y='sepal width (cm)',data=df,kind='box')
plt.show()
# dots far away in output are outliers
```

###  violinplot

#  Pair Plot

```python
sns.pairplot(data=df,hue='classes')
plt.show()
```

###  correlation:
it provides the relation between two features

* relation=0---->there is no relation  between features
* relation=1---->positive correlation
* relation=-1---->negative correation

```python
# find the correlation between two features
cor=df.corr()
cor
```

```python
sns.heatmap(cor,annot=True)
```

```python
sns.heatmap(cor,annot=True)
plt.show()
```

###  Regression plot:
Regression plot provides linear model of the dataset.
* it is used to implot for data visualization

```python
sns.lmplot(x='sepal width (cm)',y='petal length (cm)',data=df,hue='classes')
plt.show()
```
