# Groups

````python
d={"rollnumber":[101,293,374,484,909,463,788],"sname":["a","b","c","d","e","f","g"],"sage":[10,34,22,42,26,55,88],"smarks":[100,97,88,98,60,87,99],"branch":['cse','eee','ece','it','cse','mech','cse']}
d
````

````python
df=pd.DataFrame(d)
df
````

````python
g=df.groupby('branch')           # it returns object
g
````

````python
# to group it we can use  loops,or function
for i,j in g:
    print(i)                        # first way
    print(j)
````

````python
# read grouping data
g=df.groupby('branch')              # 2nd way
g
````

````python
g.get_group('cse')
````

````python
g.get_group('it')
````

````python
cse=g.get_group('cse')              # sum the cse marks
cse['smarks'].sum()
````

# Concate:

````python
d1={'city':['vizag','hyderabad','kurnool','mumbai'],'count':[100,200,300,400],'square':[20,30,40,50]}
df=pd.DataFrame(d1)
df
````

````python
d2={'city':['delhi','kerala','chennai','panjab'],'count':[500,700,800,900],'square':[60,70,80,90]}
df2=pd.DataFrame(d2)
df2
````

````python
total_df=pd.concat([df,df2])
total_df
````

````python
total_df=pd.concat([df,df2],ignore_index=True)    #  it give continue index
total_df
````

````python
total_df=pd.concat([df,df2],keys=['a','b'])
total_df
````

# merge/join

````python
import pandas as pd
d1={'sno':[199,256,356,879],'sname':['a','b','c','d'],'sage':[10,29,47,34]}
df=pd.DataFrame(d1)
df
````

````python
d2={'saddress':[199,256,389,879],'sname':['h','b','g','d'],'smarks':[80,40,88,44]}
df2=pd.DataFrame(d2)
df2
````

````python
df_merge=pd.merge(df,df2,on='sname')       # on=based on,***on*** function we merging
df_merge                                    # defaultly it is inner
````

````python
df_merge=pd.merge(df,df2,on='sname',how='inner')
df_merge
````

````python
df_merge=pd.merge(df,df2,on='sname',how='left')
df_merge
````

````python
df_merge=pd.merge(df,df2,on='sname',how='right')
df_merge
````

````python
df_merge=pd.merge(df,df2,on='sname',how='outer')
df_merge
````

````python
df_outer_join=pd.merge(df,df2,on='sname',how='outer')
df_outer_join
````

````python
df_outer_join=pd.merge(df,df2,on='sname',how='left')
df_outer_join
````
